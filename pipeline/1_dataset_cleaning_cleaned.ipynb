Strip leading/trailing whitespace and normalize Unicode (NFC)

```python
import unicodedata
def strip_and_normalize_nfc(text: str) -> str:
    if not isinstance(text, str):
        text = "" if text is None else str(text)
    text = text.strip()
    return unicodedata.normalize("NFC", text)
```

Replace URLs with <URL> and user handles with <USER>

```python
import re
URL_RE  = re.compile(r'((?:https?://|http?://|www\.)\S+)', flags=re.IGNORECASE)
USER_RE = re.compile(r'(?<!\w)@\w+')

def replace_urls_and_users(text: str) -> str:
    text = URL_RE.sub("<URL>", text)
    text = USER_RE.sub("<USER>", text)
    return text
```

Remove \n and similar, then collapse excess whitespace to single spaces

```python
WS_RE = re.compile(r"\s+")

def remove_newlines_and_collapse_ws(text: str) -> str:
    # replace any whitespace run (including \n, \t) with a single space
    return WS_RE.sub(" ", text).strip()
```

Cleaning:

```python
# 1. strip + normalize
train_data["comment_text"] = train_data["comment_text"].apply(strip_and_normalize_nfc)

# 2. replace URLs and user handles
train_data["comment_text"] = train_data["comment_text"].apply(replace_urls_and_users)

# 3. remove newlines + collapse whitespace
train_data["comment_text"] = train_data["comment_text"].apply(remove_newlines_and_collapse_ws)

# 4. drop true row duplicates across all columns
train_data = train_data.drop_duplicates().reset_index(drop=True)

train_data = train_data.rename(columns={"comment_text": "comment"})
```

```python
# 1. strip + normalize
test_data["comment_text"] = test_data["comment_text"].apply(strip_and_normalize_nfc)

# 2. replace URLs and user handles
test_data["comment_text"] = test_data["comment_text"].apply(replace_urls_and_users)

# 3. remove newlines + collapse whitespace
test_data["comment_text"] = test_data["comment_text"].apply(remove_newlines_and_collapse_ws)

# 4. drop true row duplicates across all columns
test_data = test_data.drop_duplicates().reset_index(drop=True)

test_data = test_data.rename(columns={"comment_text": "comment"})
```

```python
train_data = train_data.dropna(subset=["comment"])                       # drop NaN
train_data = train_data[train_data["comment"].str.strip().astype(bool)]        # drop empty/whitespace
```

```python
test_data = test_data.dropna(subset=["comment"])                       # drop NaN
test_data = test_data[test_data["comment"].str.strip().astype(bool)]        # drop empty/whitespace
```
