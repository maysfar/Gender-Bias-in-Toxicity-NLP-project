```python
import numpy as np
from tqdm import tqdm
from transformers import pipeline

# List of models to evaluate
model_names = [
    "moatazhamza194/tc-bert",
    "moatazhamza194/tc-bert_mask",
    "moatazhamza194/tc-bert_swap"
]

# Dictionary to hold results
fairness_results = {}

for model_name in model_names:
    print(f"\n🔎 Evaluating Counterfactual Fairness for: {model_name}")

    # --- Step 1: Load classifier ---
    classifier = pipeline("text-classification", model=model_name, top_k=None)

    # --- Step 2: Get scores from model ---
    scores = []
    for sentence, _, _ in tqdm(gender_sentence_pairs):
        out = classifier([sentence], truncation=True, max_length=128)
        score = out[0][0]["score"]   # double indexing for list-of-lists
        scores.append(score)

    scores = np.array(scores)

    # --- Step 3: Compute counterfactual fairness gaps ---
    pair_gaps = []
    pair_labels = []

    for i in range(0, len(gender_sentence_pairs), 2):  # step in pairs
        s1, y1, g1 = gender_sentence_pairs[i]
        s2, y2, g2 = gender_sentence_pairs[i+1]
        sc1 = scores[i]
        sc2 = scores[i+1]

        # absolute difference in predicted toxicity
        gap = abs(sc1 - sc2)

        pair_gaps.append(gap)
        pair_labels.append(y1)  # they share the same label (0 or 1)

    pair_gaps = np.array(pair_gaps)
    pair_labels = np.array(pair_labels)

    # --- Step 4: Summaries ---
    results = {
        "num_pairs": len(pair_gaps),
        "mean_gap": pair_gaps.mean(),
        "median_gap": np.median(pair_gaps),
        "max_gap": pair_gaps.max(),
        "non_toxic_mean_gap": pair_gaps[pair_labels==0].mean(),
        "toxic_mean_gap": pair_gaps[pair_labels==1].mean()
    }

    fairness_results[model_name] = results

    # Print summary
    print("=== Counterfactual Fairness (Gender Swap) ===")
    print(f"Number of pairs: {results['num_pairs']}")
    print(f"Mean gap: {results['mean_gap']:.4f}")
    print(f"Median gap: {results['median_gap']:.4f}")
    print(f"Max gap: {results['max_gap']:.4f}")
    print("\nBy class:")
    print(f"Non-toxic pairs (label=0) mean gap: {results['non_toxic_mean_gap']:.4f}")
    print(f"Toxic pairs (label=1) mean gap: {results['toxic_mean_gap']:.4f}")
```
